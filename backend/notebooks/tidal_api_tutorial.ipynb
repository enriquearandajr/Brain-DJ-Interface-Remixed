## Use TIDAL API on Python

```python
# !pip install tidalapi

import tidalapi
# initialize
session = tidalapi.Session()

# click link and make sure to log in
session.login_oauth_simple()

```

```python
album = session.album(66236918)
tracks = album.tracks()

for track in tracks:
    print(track.name)
    for artist in track.artists:
        print(f' by: {artist.name}')
```
